- **SkinThicknessTriceps** skin fold thickness (mm)
- **Insulin** 2-Hour serum insulin (mu U/ml)
- **BMIBody** mass index (weight in kg/(height in m)^2)
- **DiabetesPedigreeFunctionDiabetes** pedigree function

```python
#aqui se asigna los valores de los NO diabeticos que vienen en cero asigne el promedio de los rangos normales, habria que validarlo
dd = diabetes_data[(diabetes_data.Insulin == 0) & (diabetes_data.Outcome == 0)] # filtramos dataframe
features_constant = ['Insulin']
imputer_constant = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value=70, missing_values = 0))])
preprocessor = ColumnTransformer(transformers=[('imputer_constant', imputer_constant, features_constant)])
preprocessor.fit(dd)
dd = preprocessor.transform(dd)
dd
```

```python
#aqui se calcula la media de la muestra que es diabetica y tienen insulina por enciama de 0 para saber que valor asignar
ddmedia = diabetes_data.Insulin[(diabetes_data.Insulin > 0) & (diabetes_data.Outcome == 1)].mean()
print(ddmedia)
```

```python
#aqui se filtra los valores de diabeticos que vienen en cero asigne el promedio
dd2 = diabetes_data[(diabetes_data.Insulin == 0) & (diabetes_data.Outcome == 1)]
features_constant = ['Insulin']
imputer_constant = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value=ddmedia, missing_values = 0))])
preprocessor = ColumnTransformer(transformers=[('imputer_constant', imputer_constant, features_constant)])
preprocessor.fit(dd2)
dd2 = preprocessor.transform(dd2)
dd2
#print(imputer_constant)
```

```python
dd_mean = diabetes_data[(diabetes_data.Glucose == 0) | (diabetes_data.BloodPressure == 0)| (diabetes_data.BMI == 0) | (diabetes_data.SkinThickness == 0)]
features_mean = ['Glucose', 'BloodPressure','BMI', 'SkinThickness']
imputer_mean = Pipeline(steps=[('imputer', SimpleImputer(strategy='mean', missing_values = 0)),])
preprocessor = ColumnTransformer(transformers=[('imputer_constant', imputer_mean, features_mean)])
preprocessor.fit(dd_mean)
dd_mean = preprocessor.transform(dd_mean)
dd_mean
```
